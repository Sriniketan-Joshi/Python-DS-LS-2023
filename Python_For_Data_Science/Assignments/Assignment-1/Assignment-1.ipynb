```python
import pandas as pd
```

```python
census_df = pd.read_csv('NDAP_REPORT_6000.csv')
```

```python
census_df
```

```python
census_df.columns
```

```python
literacy_state_df = census_df.groupby('State')[['Population', 'Male population', 'Female population', 'Literate population ', 'Male literate population', 'Female literate population ']].sum()
```

```python
literacy_state_df
```

```python
literacy_state_df['Literacy rate'] = literacy_state_df['Literate population ']/literacy_state_df['Population'] * 100
literacy_state_df['Male literacy rate'] = literacy_state_df['Male literate population']/literacy_state_df['Male population'] * 100
literacy_state_df['Female literacy rate'] = literacy_state_df['Female literate population ']/literacy_state_df['Female population'] * 100
```

```python
literacy_state_df
```

```python
literacy_district_df = census_df.groupby('District')[['Population', 'Male population', 'Female population', 'Literate population ', 'Male literate population', 'Female literate population ']].sum()
```

```python
literacy_district_df
```

```python
literacy_district_df['Literacy rate'] = literacy_district_df['Literate population ']/literacy_district_df['Population'] * 100
literacy_district_df['Male literacy rate'] = literacy_district_df['Male literate population']/literacy_district_df['Male population'] * 100
literacy_district_df['Female literacy rate'] = literacy_district_df['Female literate population ']/literacy_district_df['Female population'] * 100
```

```python
literacy_district_df
```

```python
# State/Union Territory with highest Literacy Rate for Male population
literacy_state_df['Male literacy rate'].idxmax()
```
